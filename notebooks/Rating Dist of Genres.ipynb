```python
import pandas as pd
import numpy as np

movies_df = pd.read_csv("./ml-20m/movies.csv") #LOAD MOVIES
ratings_df = pd.read_csv("./ml-20m/ratings.csv") #LOAD RATINGS
ratings_movies_df = ratings_df.merge(movies_df, how='left', on='movieId') #MERGE MOVIES + RATINGS
```

```python
#initiating filtered dataframe for each top 5 genres 
drama_df = ratings_movies_df[ratings_movies_df['genres'].str.contains('Drama')]['rating']
comedy_df = ratings_movies_df[ratings_movies_df['genres'].str.contains('Comedy')]['rating']
thriller_df = ratings_movies_df[ratings_movies_df['genres'].str.contains('Thriller')]['rating']
romance_df = ratings_movies_df[ratings_movies_df['genres'].str.contains('Romance')]['rating']
action_df = ratings_movies_df[ratings_movies_df['genres'].str.contains('Action')]['rating']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.hist([drama_df.values, comedy_df.values, thriller_df.values, romance_df.values, action_df.values],
       rwidth=0.6,
       align='mid')
plt.title("Movie Rating Distributions by Genres")
plt.xlabel("Rating")
plt.ylabel("Distribution")
plt.ticklabel_format(style='plain')
plt.legend(['drama', 'comedy', 'thriller', 'romance', 'action'])
plt.show();

#ratings between genres is quite evenly distributed with highest count
#between 3-4
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(drama_df.values, bins=10, alpha=0.3)
sns.histplot(comedy_df.values, bins=10, alpha=0.3)
sns.histplot(thriller_df.values, bins=10, alpha=0.3)
sns.histplot(romance_df.values, bins=10, alpha=0.3)
sns.histplot(action_df.values, bins=10, alpha=0.3)
plt.legend(['drama', 'comedy', 'thriller', 'romance', 'action'])
plt.ticklabel_format(style='plain')
plt.show()
```
